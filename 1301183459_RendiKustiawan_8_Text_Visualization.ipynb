```python
logobibit_mask = np.array(Image.open('logobibit.png'))
logobibit_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
transformed_logobibit_mask = np.ndarray((logobibit_mask.shape[0],logobibit_mask.shape[1]), np.int32)

for i in range(len(logobibit_mask)):
    transformed_logobibit_mask[i] = list(map(transform_format, logobibit_mask[i]))
transformed_logobibit_mask
```

```python
Image.open('logobibit.png')
```

```python
# Generate Word
wc = WordCloud(
    background_color="white",
    max_words=1000,
    mask=transformed_logobibit_mask,
    stopwords=stopwords,
    mode='RGBA'
).generate(article)

# Show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
