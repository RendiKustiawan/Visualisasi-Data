### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

```python
#Hang-seng
df_hangseng = pd.read_csv('/content/drive/MyDrive/Hang_Seng.csv')
df_hangseng
```

```python
df_nasdaq = pd.read_csv('/content/drive/MyDrive/Nasdaq.csv')
df_nasdaq
```

```python
df_nikkei = pd.read_csv('/content/drive/MyDrive/Nikkei.csv')
df_nikkei
```

```python
# type your code here
fig = plt.figure()
fig, (ax0, ax1, ax2) = plt.subplots(1, 3)

#Hang-Seng
df_hangseng.index = df_hangseng["Date"]
df_hangseng = df_hangseng.loc['2019-01-01':'2019-31-12']

df_hangseng["Day_Perc_Change"] = df_hangseng["Adj Close"].pct_change()*100
df_hangseng.dropna(inplace=True)
df_hangseng["Trend"]= df_hangseng["Day_Perc_Change"].apply(lambda x:trend(x))

hangseng_pie_data = df_hangseng.groupby('Trend')
pie_label = sorted([i for i in df_hangseng.loc[:, 'Trend'].unique()])

explode_list = [0.1,0, 0, 0, 0, 0]
colors_list = [ '#ff0a54', '#ff477e', '#ff7096', '#ff85a1', '#fbb1bd', '#f7cad0']
hangseng_pie_data['Date'].count().plot(
        kind = 'pie',
        ax =ax0,
        autopct = '%1.1f%%',
        shadow =True,
        explode = explode_list,
        colors = colors_list,
        figsize = (30,8)
        )
ax0.set_title('Hang-Seng Trend')
ax0.set_ylabel('')

#Nasdaq

df_nasdaq.index = df_nasdaq["Date"]
df_nasdaq = df_nasdaq.loc['2019-01-01':'2019-31-12']

df_nasdaq["Day_Perc_Change"] = df_nasdaq["Adj Close"].pct_change()*100
df_nasdaq.dropna(inplace=True)
df_nasdaq["Trend"]= df_nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))

nasdaq_pie_data = df_nasdaq.groupby('Trend')
pie_label = sorted([i for i in df_nasdaq.loc[:, 'Trend'].unique()])

explode_list = [0.1,0, 0, 0, 0, 0]
colors_list = [ '#006400', '#007200', '#008000', '#38b000', '#9ef01a', '#ccff33']
nasdaq_pie_data['Date'].count().plot(
        kind = 'pie',
        ax =ax1,
        autopct = '%1.1f%%',
        shadow =True,
        # explode = explode_list,
        colors = colors_list
        )
ax1.set_title('Nasdaq Trend')
ax1.set_ylabel('')

#Nikkei
df_nikkei.index = df_nikkei["Date"]
df_nikkei = df_nikkei.loc['2019-01-01':'2019-31-12']

df_nikkei["Day_Perc_Change"] = df_nikkei["Adj Close"].pct_change()*100
df_nikkei.dropna(inplace=True)
df_nikkei["Trend"]= df_nikkei["Day_Perc_Change"].apply(lambda x:trend(x))

nikkei_pie_data = df_nikkei.groupby('Trend')
pie_label = sorted([i for i in df_nikkei.loc[:, 'Trend'].unique()])

explode_list = [0.1,0, 0, 0, 0, 0]
colors_list = [ '#ff7b00', '#ff9500', '#ffa200', '#ffc300', '#ffdd00', '#ffea00']
nikkei_pie_data['Date'].count().plot(
        kind = 'pie',
        ax =ax2,
        autopct = '%1.1f%%',
        shadow =True,
        explode = explode_list,
        colors = colors_list
        )
ax2.set_title('Nikkei Trend')
ax2.set_ylabel('')

plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
df_combined = pd.concat([df_hangseng['Adj Close'], df_nasdaq['Adj Close'], df_nikkei['Adj Close']], axis=1)
df_combined.columns = ['Hang-Seng', 'Nasdaq', 'Nikkei']

df_combined.dropna(inplace = True, axis = 0)

pct_chg_df = df_combined.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

volatility = pct_chg_df[['Hang-Seng', 'Nasdaq', 'Nikkei']].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 10))


plt.title('Trend Hangseng, Nasdaq dan Nikkei pada tahun 2019')
plt.show()
```

#### Penjelasan 
Dari tiga trend diatas kita bisa lihat bahwa nasdaq memiliki volatilitas yang tinggi dibandingkan dengan yang hangseng dan nikkei karena berubah secara dramatis dalam kurun waktu yang cepat. Dan volatilitas dari Hang-seng dan Nikkei memiliki valatilitas yang rendah dibandingkan dengan nasdaq karena nilai saham cenderung lebih stabil. Dapat disimpulkan bahwa ketiga saham tersebut memiliki volatilitas yang cukup baik karena pergerakan cukup cepat tetapi dalam kasus ini yang memiliki volatilitas yang paling tinggi adalah nasdaq dan volatilitas menengah/rendah adalah hangseng dan nikkei.
